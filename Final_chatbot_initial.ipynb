```python
import pandas as pd
chatbot_data = pd.read_excel('/content/drive/MyDrive/프로젝트/chatbot_demo.xlsx')
chatbot_data
```

```python
chat_dict = {}
row = 0
for rule in chatbot_data['rule']:
    chat_dict[row] = rule.split('|')
    row += 1
print(chat_dict)
```

```python
def chat(request):
    for k, v in chat_dict.items():
        chat_flag = False
        for word in v:
            if word in request:
                chat_flag = True
            else: 
                chat_flag = False
                break
        if chat_flag:
            return chatbot_data['response'][k]
    return "무슨 말인지 모르겠어요."
```

```python
def chat(request):
    for k, v in chat_dict.items():
        index = -1
        for word in v:
            try:
                if index == -1:
                    index = request.index(word)
                else:
                    if index < request.index(word, index):
                        index = request.index(word, index)
                    else:
                        index = -1
                        break
            except ValueError:
                index = -1
                break
        if index > -1:
            return chatbot_data['response'][k]
    return '무슨 말인지 모르겠어요.'
```

```python
while True:
    req = input('무엇이 궁금한가요?: ')
    if req == 'exit':
        break
    else:
        print(chat(req))
```
